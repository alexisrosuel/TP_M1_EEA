### On va créer un signal de type sinusoidal numérique, qu'on va ensuite écouter.

```python
fs = 44100 # fq d'échantillonage
T = 2 # duree du signal en secondes
fq = 440
points = np.arange(0,T * fs)/fs
y = 20*np.cos(2*np.pi*fq * points)
```

```python
from IPython.display import Audio
Audio(10*y, rate = 44100)
```

### On représente une partie du signal, par exemple ci-dessous, 5 maximums sont détectés tous les centièmes de secondes environ. Ce qui signifie bien qu'on a une fréquence de l'ordre de 500 Hz.

```python
### on représente une partie du signal
pl.plot(points[0:1000],y[0:1000])
pl.show()
```

```python
import scipy.fftpack as sf
slices = slice(0,10000,1)
### on appelle la transformée de Fourier rapide sur le signal y.
fft_y = sf.fft(y)
pl.plot(points[slices]*fs,fft_y[slices])
```

### On observe un pic à une certaine position. On détecte cette position en utilisant la fonction argmax et on la traduit en fréquence.

```python
a =  np.argmax(fft_y[slices])
estimation_fq = float(a)/len(fft_y)*fs
message = "la fréquence pour laquelle la transformée de Fourier est maximale est {} Hz".format(estimation_fq)
print(message)
```

### On retrouve bien, dans ce cas, la fréquence du son qu'on a créé. On propose de faire la même chose pour un fichier son, qui est l'enregistrement d'une note de musique. On importe un fichier son, on lit la fréquence d'échantillonnage. Le signal étant stéréo, il y a, dans le même vecteur les deux sons concaténés: 

```python
from scipy.io import wavfile
fs, data = wavfile.read('piano.wav')
print("la frequence est ",fs," donc 1 seconde du signal correspond à ",fs, " points")
print("la longueur du signal numerique mono est ",len(data)//2)

## On remet le signal à la dimension du "vrai" son.
data_stereo = np.reshape(data,(np.size(data)//2,2))
signal_mono = 0.5 * np.sum(data_stereo,axis = 1)
print("la longueur du signal mono en secondes est ", len(signal_mono)//fs)
```

### On peut écouter le signal mono (c'est-à-dire une seule sortie pour le son).

```python
from IPython.display import Audio
Audio(signal_mono, rate = 44100)
```

### On représente le signal en temporel, c'est-à-dire en fonction du temps.

```python
pl.plot(np.arange(len(signal_mono))/fs,signal_mono)
pl.show()
```

### On représente le signal en Fourier en utilisant la FFT.

```python
### on importe la librairie codant la fft.
import scipy.fftpack as sf
### on doit faire un shift du signal pour avoir la fréquence nulle au milieu du graphique.
a = sf.fftshift(sf.fft(signal_mono))
x = range(len(signal_mono)//2-len(signal_mono),len(signal_mono)//2)
x = np.asarray(x,dtype=np.float32)/10000
pl.plot(x,(np.abs(a)))
pl.show()
```

### On introduit la fft sur une sous-partie du signal.

```python
## On introduit une fenetre sur laquelle on va calculer la fft.
import scipy.signal as sig

start = 44100 * 2
length = int(44100 * 0.5)
print(length)### on considère un nombre d'échantillons d'une durée équivalente de 1.5 seconde
slices = slice(start,start+length,1)
window = sig.hamming(length)
print(np.size(window))

pl.plot(window)
pl.title("Hamming window")
pl.ylabel("Amplitude")
pl.xlabel("Sample")
pl.show()
```

```python
#pl.plot(signal_mono[70000:70000 + 66150])
pl.plot(signal_mono[slices])
pl.show()
fft_sig = sf.fft(window * signal_mono[slices])
a = sf.fftshift(fft_sig)
x = range(len(signal_mono[slices])//2-len(signal_mono[slices]),len(signal_mono[slices])//2)
x = np.asarray(x,dtype=np.float32)/10000
pl.plot(x,(np.abs(a)))
pl.show()
```

### On va donc sélectionner la fréquence donnant l'amplitude maximale de la FFT.

```python
temp = np.abs(fft_sig[0:length//2])
estimation = np.argmax(temp)/(length)*fs
message = "La fréquence pour laquelle la transformée de Fourier est maximale est {} Hz".format(estimation)
print(message)
```

### On effectue la même expérience sur un signal sonore différent. Ce signal est une conçonne, un "s".

```python
from scipy.io import wavfile
fs, data = wavfile.read('sss.wav')
print("la frequence est ",fs," donc 1 seconde du signal correspond à ",fs, " points")
print("la longueur du signal numerique mono est ",len(data)//2)
print("la forme du signal est "+str(np.shape(data)))
## On remet le signal à la dimension du "vrai" son.
data_stereo = np.reshape(data,(np.size(data)//2,2))
signal_mono = 0.5 * np.sum(data_stereo,axis = 1)
print("la longueur du signal mono en secondes est ", len(signal_mono)//fs)
```

```python
from IPython.display import Audio
Audio(signal_mono, rate = 44100)
```

```python
pl.plot(np.arange(len(signal_mono))/fs,signal_mono)
pl.show()
```

### La transformée de Fourier de ce signal ne présente pas une harmonique importante en particulier mais un paquet d'harmoniques regroupées sur des intervalles en fréquentiel.

```python
a = sf.fftshift(sf.fft(signal_mono))
x = range(len(signal_mono)//2-len(signal_mono),len(signal_mono)//2)
x = np.asarray(x,dtype=np.float32)/10000
pl.plot(x,(np.abs(a)))
pl.show()
```

```python
temp = np.abs(a)
estimation = np.argmax(temp)/(len(temp))*fs
message = "La fréquence pour laquelle la transformée de Fourier est maximale est {} Hz".format(estimation)
print(message)
```

## Le théorème de Shannon

```python
## On définit un signal
def mySignal(t):
    fq = 0.9
    fq2 = 0.3
    return np.sin(2*np.pi * fq*t) + 0.2*np.cos(2*np.pi * fq2*t)
x = np.linspace(0,2*np.pi,1000)
pl.plot(x,mySignal(x))
pl.show()
```

### Exercice: 
+ Choisissez un temps d'interpolation $T_e$ et montrer la reconstruction du signal en utilisant une interpolation linéaire, c'est-à-dire des segments entre les points de la courbe.
+ Implémenter la formule d'interpolation de Shannon et comparer avec l'interpolation linéaire; changer de forme de signal.

```python
# linéaire
for Te in [0.1,0.5,1]:
    plt.figure()
    indices = np.arange(0,2*np.pi,Te)
    echantillon = mySignal(indices)
    plt.plot(indices, echantillon)

    x = np.linspace(0,2*np.pi,1000)
    pl.plot(x,mySignal(x))
    plt.title("Te=%s" % Te)
```

```python
# shannon : reconstruction parfait dès que Te<0.5
def reconstruction(t, Te=1):
    N_range = np.arange(-1000,1000,1)
    return np.sum(mySignal(N_range*Te) * np.sinc(1/Te*(t-N_range*Te)))

for Te in np.arange(0,1,0.1):
    plt.figure()
    x_range = np.arange(0,2*np.pi,0.01)
    plt.plot(x_range, [reconstruction(xi,Te=Te) for xi in x_range])

    x = np.linspace(0,2*np.pi,1000)
    pl.plot(x,mySignal(x))
    plt.title("Te=%s" % Te)
```
